## Loading the base FHIR spec
Before we can parse new profiles we need to load the base FHIR spec.

```python
from generate_models import load_fhirspec


fhirspec = load_fhirspec()
```

## Building a new profile
Using the loaded fhir spec and the definition of a new profile we will generate a profile object.

```python
profile_path = Path('../profiles/zib-LaboratoryTestResult-DiagnosticReport.json')

with profile_path.open('r') as f:
    profile_dict = json.load(f)
    
    
profile = FHIRStructureDefinition(fhirspec, profile_dict)
profile
```

```python
profile.process_profile()
```
